#### 코드 설명

(1) os.makedirs : 디렉토리 생성 (os.mkdir와 달리 원하는 만큼 디렉토리 생성 가능)
  - [exist_ok = True : 해당 디렉토리가 기존에 존재하면 에러발생 없이 넘어가고, 없을 경우에만 생성]

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
import time
import urllib.request
import os

pumjong = {"한이음": ["뚱한 표정"]}


def crawling(target_name):
    
    SCROLL_PAUSE_TIME = 2  
    count = 0
    
    # =============================================================================
    # 구글 이미지 검색 접속 및 검색어 입력
    # =============================================================================
    driver.get("https://www.google.co.kr/imghp?hl=ko&tab=wi&ogbl") # 크롤링할 사이트 주소 입력
    elem = driver.find_element_by_name("q") # 요소의 이름이 "q"인 것을 찾아서 변수 할당(구글 이미지 사이트의 입력창)
    elem.send_keys(target_name) # 크롤링을 할 사진 제목명 입력
    elem.send_keys(Keys.RETURN) # Enter kEY를 누른다.
    


    # =============================================================================
    # 스크롤
    # =============================================================================
    print(target_name +' 스크롤 중 .............')
    
    elem_body =  driver.find_element_by_tag_name("body")
    for i in range(60):
        elem_body.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
    
    try:
        driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div[1]/div[4]/div[2]/input').click()
        for i in range(60):
            elem_body.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.1)
    except:
        pass
    
    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd") # image를 가져오기 
    print(target_name +' 찾은 이미지 개수:',len(images))

    for image in images:            
        try:
            image.click() # 사진을 클릭 
            time.sleep(2) # 대기 
            url =  driver.find_element_by_xpath('/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[2]/div/a/img').get_attribute("src")
            print(url)
            urllib.request.urlretrieve(url, str(count) + ".jpg")  # 사진 저장 
        except:
            pass
        count = count + 1
        
driver = webdriver.Chrome()
        
for key in pumjong:
    os.makedirs(key, exist_ok=True)  # 디렉토리 생성 
    os.chdir(key) # 디렉터리 위치 변경하기
    for val in pumjong[key]:
        os.makedirs(val, exist_ok=True) # 디렉토리 생성 
        os.chdir(val) # 디렉터리 위치 변경하기
        crawling(val)
        os.chdir('..') 
    os.chdir('..')
driver.close()
```
